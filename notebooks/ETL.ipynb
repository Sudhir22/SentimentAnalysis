```python
import gzip
import pandas as pd
```

```python
"""
Code from http://jmcauley.ucsd.edu/data/amazon/
"""
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('../data/reviews_Office_Products_5.json.gz')
```
